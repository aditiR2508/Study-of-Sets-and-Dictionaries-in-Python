```python
#set is a collection which is unordered-->no fixed order, hence no indicing possible
#sets are used to store multiple items in a single variable
#set is a sub-type of dictionary
#output sequence is random


#creation of a set
guitar = {"bass", "acoustic", "electric"}
print(guitar)
```

```python
#sets do not allow duplicate items; are unordered and unchangeable
guitar = {"bass", "acoustic", "electric", "bass"}
print(guitar)
```

```python
# True and 1 are considered same value in sets; 0 and False are considered same
guitar = {"bass", "acoustic", "electric", True, 1, False, 0}
print(guitar)
```

```python
#set supports all types of data types like in lists
forests={"rain", "evergreen", 0.785, 1, True, False, 58}
forests       #prints set
```

```python
#searching for item in set
forests={"rain", "evergreen", 0.785, 1, True, False, 58}
print("rain" in forests)
print("pine" in forests)
```

```python
#once you create a set, you cannot replace any of its items, however adding items to the set is possible
forests={"rain", "evergreen", 0.785, 1, True, False, 58}
forests.add("pine")
print(forests)
```

```python
#removing items
forests={"rain", "evergreen", 0.785, 1, True, False, 58}
forests.remove("rain")
forests
```

```python
#Set Operations
AC={1, 23, 45, 67}
DC={0, 1, 2, 3, 4, 5, 6, 7}

print("Union:", AC | DC)
print("Intersection: ", AC & DC)
print("Difference: ", AC-DC)
print("Symmetric Difference: ", AC ^ DC)
```

```python
#Frozen set
forests= frozenset({"rain", "evergreen", 0.785, 1, True, False, 58})
print(forests)
print(type(forests))
```

```python
#Addition or Removal of elements in Frozensets--->NOT possible
forests= frozenset({"rain", "evergreen", 0.785, 1, True, False, 58})
forests.add("pine")
forests.remove("rain")
print(forests)
```

# Dictionaries in Python

```python
# A Dictionary is a collection which is changeable(we can make changes to it), ordered and does not allow duplicates.

# Creating a Dictionary
car = {"Brand": "Volkswagen", "Model": "Jetta", "Year": 2017}
print(car)
```

```python
#Dictionary does not allow duplicate values and will take the latest value provided to it
car = {"Brand": "Volkswagen", "Model": "Jetta", "Model": "Tiguan", "Year": 2017}
print(car)
```

```python
#length of a dictionary
car = {"Brand": "Volkswagen", "Model": "Jetta", "Year": 2017}
print(len(car))
```

```python
#Changing or Adding key-value to a dictionary
car = {"Brand": "Volkswagen", "Model": "Jetta", "Year": 2017}
car["Model"] = "Tiguan"
print(car)
car["Colour"] = "Midnight Blue"
print(car)
x = car.values()
print(x)
```

```python
#Retrieving Values
student_data = {"PRN": "25578090789", "Name": "Amar Anthony", "Branch": "EnTC", "Division": "A", "Home City": "Bhatinda"}
print("Name:", student_data["Name"])
```

```python
#Adding & Updating Elements
student_data["Year"] = "Third"
student_data["Name"] = "Ajay Anthony"
print(student_data)
```

```python
#Removing Elements---->using pop() function
student_data = {"PRN": "25578090789", "Name": "Amar Anthony", "Branch": "EnTC", "Division": "A", "Home City": "Bhatinda"}

print("Before Removal: ", student_data)

student_data.pop("Home City")
print("After Removal: ", student_data)
```

```python
#Problem Statement 1: A given Dictionary stores product names and their prices, python program to update the price of a given product.
grocery = {"Pulses": 70, "Rice": 50, "Flour": 45, "Vegetables": 125}

print("Original Products & Prices: ", grocery)  #original price

grocery["Vegetables"] = 175
print("Updated Products & Prices: ", grocery)   #updated price
```

```python
#Problem Statement 2: A dictionary contains student names and marks:
### find a student's marks using the name entered.
### if the student is not found, display a suitable message.

marks = {"Amar": 45, "Ajit": 47, "Rahul": 50, "Ajay": 38, "Karan": 42}

x = input("Enter a Student's Name: ")

print(marks.get(x, "The student name entered does not exist.")) ##using get() function to fetch the marks of the student whose name is entered
```

```python
#Problem Statement 3: A dictionary stores usernames and passwords, write a python program to validate user login.
user_credentials = {"amar@13": "1278", "kabir#24": "4534", "$veer$": "1729"}
user_name = input("Enter Username: ")
password = input("Enter Password: ")
if user_credentials.get(user_name) == password:
  print("User Login was Successful!")
else:
  print("Login Failed!!!")
```

```python
#Problem Statement 4: A Dictionary stores student names and marks, to find the student who scored the highest marks.
student_details = {"Ajit": 34, "Amar": 46, "Matthew": 42, "Karan": 49}
max_marks = max(student_details, key=student_details.get)
print("Student having highest marks: ", max_marks)
print("Marks: ", student_details[max_marks])
```
